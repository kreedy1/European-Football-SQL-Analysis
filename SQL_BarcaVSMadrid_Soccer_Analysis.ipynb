```python
detailed_matches = pd.read_sql("""
SELECT Match.id, 
Country.name AS country_name, 
League.name AS league_name, 
season, 
date,
HT.team_long_name AS  home_team,
AT.team_long_name AS away_team,
home_team_goal, 
away_team_goal                                        
FROM Match
LEFT JOIN Country on Country.id = Match.country_id
LEFT JOIN League on League.id = Match.league_id
LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
WHERE season = '2015/2016'
AND country_name = 'Spain'
ORDER by date
LIMIT 10;""", conn)
detailed_matches
```

```python
barc = pd.read_sql("""
SELECT m.date,
HT.team_long_name AS  home_team,
AT.team_long_name AS away_team,
home_team_goal, 
away_team_goal, 
CASE WHEN m.home_team_api_id = 8634 and m.home_team_goal > m.away_team_goal THEN 'Barca win'
WHEN m.home_team_api_id = 8634 and  m.home_team_goal < m.away_team_goal THEN 'Barca loss'
WHEN m.away_team_api_id = 8634 and m.home_team_goal < m.away_team_goal THEN 'Barca win'
WHEN m.away_team_api_id = 8634 and m.home_team_goal > m.away_team_goal THEN 'Barca loss'
ELSE 'Tie' END AS outcome
FROM match m
LEFT JOIN Team AS HT on HT.team_api_id = m.home_team_api_id
LEFT JOIN Team AS AT on AT.team_api_id = m.away_team_api_id
WHERE m.home_team_api_id = 8634 or m.away_team_api_id = 8634
Order by 1;""", conn)
barc
```

Barcelona Seasons Overview

```python
barc_season = pd.read_sql("""
SELECT
    m.season AS Season,
    SUM(CASE 
        WHEN m.home_team_api_id = 8634 AND m.home_team_goal > m.away_team_goal THEN 1
        WHEN m.away_team_api_id = 8634 AND m.home_team_goal < m.away_team_goal THEN 1
        ELSE 0
    END) AS Win,
    SUM(CASE
        WHEN m.home_team_api_id = 8634 AND m.home_team_goal = m.away_team_goal THEN 1
        WHEN m.away_team_api_id = 8634 AND m.home_team_goal = m.away_team_goal THEN 1
        ELSE 0
    END) AS Tie,
    SUM(CASE 
        WHEN m.home_team_api_id = 8634 AND m.home_team_goal < m.away_team_goal THEN 1
        WHEN m.away_team_api_id = 8634 AND m.home_team_goal > m.away_team_goal THEN 1
        ELSE 0
    END) AS Loss,
    COUNT(*) AS Matches_Played  -- Counting the number of rows in the group
FROM 
    match m
LEFT JOIN 
    Team AS HT ON HT.team_api_id = m.home_team_api_id
LEFT JOIN 
    Team AS AT ON AT.team_api_id = m.away_team_api_id
WHERE 
    m.home_team_api_id = 8634 OR m.away_team_api_id = 8634
GROUP BY 
    Season;""", conn)
barc_season
```

Barcelona Win Rate at Home and Away

Barcelona High Score Matches Using CTEs

```python
Barcelona_High_Score_Match = pd.read_sql("""
WITH Barcelona_High_Score AS (
    SELECT 
        Match.id, 
        Match.season, 
        DATE(Match.date),
        HT.team_long_name AS home_team,
        AT.team_long_name AS away_team,
        home_team_goal, 
        away_team_goal,
        home_team_goal + away_team_goal AS total_goals
    FROM 
        Match
    LEFT JOIN 
        Team AS HT ON HT.team_api_id = Match.home_team_api_id
    LEFT JOIN 
        Team AS AT ON AT.team_api_id = Match.away_team_api_id
    WHERE 
        (Match.home_team_api_id = 8634 OR Match.away_team_api_id = 8634)
        AND (home_team_goal + away_team_goal) > 4
)
SELECT 
    *
FROM 
    Barcelona_High_Score;""", conn)
Barcelona_High_Score_Match
```

Before looking into the data for Barcelona and Real Madrid, lets take a larger overview by looking into the different leagues:

```python
Avg_League_Goals = pd.read_sql("""
SELECT 
League.name AS league_name,
ROUND(AVG(home_team_goal + away_team_goal),2) as avg_goals
FROM Match
LEFT JOIN League on League.id = Match.league_id
GROUP BY league_name                      
;""", conn)
Avg_League_Goals
```

```python
Avg_league_Goals_rank = pd.read_sql("""
SELECT 
League.name AS league_name,
ROUND(AVG(home_team_goal + away_team_goal),2) as avg_goals,
RANK() OVER(ORDER BY AVG(home_team_goal + away_team_goal)DESC) AS league_rank
FROM Match
LEFT JOIN League on League.id = Match.league_id
GROUP BY league_name                      
;""", conn)
Avg_league_Goals_rank
```

Its interesting to see this ranking. As a passionate football fan, I expected the smaller European leagues to have more goals and the top 5 leagues to be at the bottom of the ranks.

Going back to the two teams I am analyzing, lets take a look at firstly Barcelonas matches with average home and away goals on aggregate:

```python

barca_matches_average = pd.read_sql("""
SELECT
SUBSTRING(Match.date,1,11) AS date,
season,
HT.team_long_name AS  home_team,
AT.team_long_name AS away_team,
home_team_goal, 
away_team_goal,
CASE WHEN  Match.home_team_api_id = 8634 THEN 'home'
ELSE 'away' END as Home_Or_Away,
AVG (home_team_goal) OVER (partition by season,strftime('%m', Match.date)) as season_month_home_avg,
AVG (away_team_goal) OVER (partition by season,strftime('%m', Match.date)) as season_month_away_avg
FROM Match
LEFT JOIN League 
ON League.id = Match.league_id
LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
WHERE Match.home_team_api_id = 8634 or Match.away_team_api_id = 8634;""", conn)
barca_matches_average
```

Running Total Using Window Function for Barcelona and Real Madrid

```python
barca_running_total = pd.read_sql("""
SELECT 
SUBSTR(Match.date, 1, 11) AS date,
season,
HT.team_long_name AS  home_team,
home_team_goal,
SUM(home_team_goal) OVER(order by match.date ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) As Running_Total
FROM Match
LEFT JOIN Team AS HT 
on HT.team_api_id = Match.home_team_api_id
WHERE Match.home_team_api_id = 8634 
                                ;""", conn)
barca_running_total
```

```python
rmadrid_running_total = pd.read_sql("""
SELECT 
SUBSTR(Match.date, 1, 11) AS date,
season,
HT.team_long_name AS  home_team,
home_team_goal,
SUM(home_team_goal) OVER(order by match.date ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) As Running_Total
FROM Match
LEFT JOIN Team AS HT 
on HT.team_api_id = Match.home_team_api_id
WHERE Match.home_team_api_id = 8633
                                ;""", conn)
rmadrid_running_total
```

In the 7 year dataset, Real Madrid has scored 10 more goals than Barcelona which is only a 0.5% difference amongst the total 1000 goals they have scored combined!
